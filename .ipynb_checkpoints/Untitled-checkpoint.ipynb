```python
import pandas as pd
import numpy as np

import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, classification_report
```

```python
file_path = r"C:\Users\FC\Desktop\Bazia\Customer Churn.xlsx"
df = pd.read_excel(file_path)

df.head()
```

```python
# Clean column names: lower + replace spaces with underscore
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df.columns.tolist()
```

```python
target = "churn"

service_quality_vars = ["call__failure", "complains"]
pricing_vars = ["charge__amount", "tariff_plan"]
engagement_vars = ["subscription__length", "frequency_of_use", "frequency_of_sms"]

features = service_quality_vars + pricing_vars + engagement_vars
```

```python
model_df = df[[target] + features].copy()
model_df = model_df.dropna()
model_df.head()
```

```python
model_df_encoded = pd.get_dummies(model_df, columns=["tariff_plan"], drop_first=True)
model_df_encoded.head()
```

```python
# Basic summary
print("Rows, Columns:", model_df_encoded.shape)

# Churn rate
churn_rate = model_df_encoded["churn"].mean()
print("Churn Rate (%):", round(churn_rate*100, 2))

# Descriptive statistics
desc = model_df_encoded.describe().T
desc
```

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, classification_report

X = model_df_encoded.drop(columns=["churn"]).copy()
y = model_df_encoded["churn"].copy().astype(int)

# Make sure numeric
X = X.apply(pd.to_numeric, errors="coerce").replace([np.inf, -np.inf], np.nan)
data = pd.concat([y, X], axis=1).dropna()
y = data["churn"].astype(int)
X = data.drop(columns=["churn"])

# Convert to numpy (statsmodels safe)
X_np = sm.add_constant(X).astype(float).values
y_np = y.values.astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X_np, y_np, test_size=0.2, random_state=42, stratify=y_np
)

logit = sm.Logit(y_train, X_train).fit()
print(logit.summary())
```

```python
y_prob = logit.predict(X_test)
y_pred = (y_prob >= 0.5).astype(int)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Feature names (including constant)
feature_names = ["const"] + list(X.columns)

params = pd.Series(logit.params, index=feature_names)
odds_ratios = np.exp(params)

pvals = pd.Series(logit.pvalues, index=feature_names)

results_table = pd.DataFrame({
    "coef": params,
    "odds_ratio": odds_ratios,
    "p_value": pvals
}).sort_values("p_value")

results_table
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr = model_df_encoded.corr(numeric_only=True)
corr
plt.figure(figsize=(10,8))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Matrix of Variables")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot heatmap
plt.figure(figsize=(5,4))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Not Churned", "Churned"],
    yticklabels=["Not Churned", "Churned"]
)

plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Confusion Matrix for Customer Churn Prediction")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.3f})")
plt.plot([0, 1], [0, 1], linestyle="--", label="Random Guess")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve for Churn Prediction Model")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

feature_names = ["const"] + list(X.columns)
params = pd.Series(logit.params, index=feature_names)
odds_ratios = np.exp(params)

# Remove constant
odds_ratios = odds_ratios.drop("const").sort_values()

plt.figure(figsize=(7,5))
odds_ratios.plot(kind="barh")
plt.axvline(1, linestyle="--")  # OR = 1 baseline

plt.xlabel("Odds Ratio")
plt.title("Odds Ratios of Predictors (Logistic Regression)")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Build a small dataframe for plotting
plot_df = pd.DataFrame({
    "Actual_Churn": y_test,
    "Predicted_Probability": y_prob
})

plt.figure(figsize=(7,5))
sns.kdeplot(
    data=plot_df,
    x="Predicted_Probability",
    hue="Actual_Churn",
    fill=True,
    common_norm=False
)

plt.title("Predicted Probability Distribution by Actual Churn Class")
plt.xlabel("Predicted Probability of Churn")
plt.tight_layout()
plt.show()
```
